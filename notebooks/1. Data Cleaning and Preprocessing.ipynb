```python
book_tags = book_tags.merge(tags, on="tag_id", how="left")
```

```python
book_tags = book_tags.sort_values(
    "count",
    ascending=False
)
```

```python
book_tags = book_tags.drop_duplicates(
    subset="goodreads_book_id"
)
```

```python
book_tags = book_tags[
    [
        "goodreads_book_id",
        "tag_name"
    ]
]
```

```python
book_tags.head()
```

```python
book_tags.rename(
    columns={
        "goodreads_book_id": "Book_ID",
        "tag_name": "Genre"
    },
    inplace=True
)
```

```python
book_tags.head()
```

```python
genre_mapping = {

    # Fantasy
    "fantasy": "Fantasy",
    "magic": "Fantasy",
    "dragons": "Fantasy",
    "wizard": "Fantasy",
    "epic-fantasy": "Fantasy",
    "urban-fantasy": "Fantasy",

    # Romance
    "romance": "Romance",
    "love": "Romance",
    "chick-lit": "Romance",
    "contemporary-romance": "Romance",

    # Mystery
    "mystery": "Mystery",
    "detective": "Mystery",
    "crime": "Mystery",
    "murder": "Mystery",

    # Thriller
    "thriller": "Thriller",
    "suspense": "Thriller",

    # Horror
    "horror": "Horror",
    "paranormal": "Horror",
    "ghosts": "Horror",
    "vampires": "Horror",

    # Science Fiction
    "science-fiction": "Science Fiction",
    "sci-fi": "Science Fiction",
    "dystopia": "Science Fiction",
    "post-apocalyptic": "Science Fiction",

    # Historical
    "historical-fiction": "Historical",
    "history": "Historical",
    "historical": "Historical",

    # Biography
    "biography": "Biography",
    "memoir": "Biography",
    "autobiography": "Biography",

    # Young Adult
    "young-adult": "Young Adult",
    "ya": "Young Adult",

    # Classics
    "classics": "Classics",
    "classic": "Classics",

    # Self Help
    "self-help": "Self Help",
    "personal-development": "Self Help",

    # Business
    "business": "Business",
    "finance": "Business",
    "economics": "Business",

    # Programming
    "programming": "Programming",
    "computer-science": "Programming"
}
```

```python
book_tags["Genre"] = book_tags["Genre"].str.lower()

book_tags["Genre"] = book_tags["Genre"].map(genre_mapping)
```

```python
book_tags = book_tags.dropna()
```

```python
book_tags.head()
```

```python
books = books.merge(
    book_tags,
    on="Book_ID",
    how="left"
)
```

```python
books.head()
```

```python
books = books.dropna(subset=["Genre"])
```
